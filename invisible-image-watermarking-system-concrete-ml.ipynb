```python
# Step 1: Install necessary libraries
!pip install numpy opencv-python matplotlib

# Step 2: Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Step 3: Create a synthetic image (32x32)
def create_synthetic_image():
    # Create a simple gradient image
    image = np.zeros((32, 32), dtype=np.uint8)
    for i in range(32):
        for j in range(32):
            image[i, j] = (i * 8) % 256  # Simple gradient
    return image

# Step 4: Display image
def display_image(image):
    plt.imshow(image, cmap='gray')
    plt.axis('off')
    plt.show()

# Step 5: Watermarking function
def embed_watermark(image, watermark):
    watermark = cv2.resize(watermark, (32, 32))
    watermarked_image = image.copy()
    watermarked_image = (watermarked_image & 0xFE) | watermark
    return watermarked_image

# Step 6: Example usage
image = create_synthetic_image()  # Create a synthetic image
display_image(image)

# Create a simple binary watermark
watermark = np.zeros((32, 32), dtype=np.uint8)
watermark[10:20, 10:20] = 1  # A simple square watermark

# Embed the watermark
watermarked_image = embed_watermark(image, watermark)
display_image(watermarked_image)

# Extract watermark
def extract_watermark(watermarked_image):
    extracted_watermark = watermarked_image & 0x01  # Extract LSB
    return extracted_watermark

# Extract the watermark
extracted_watermark = extract_watermark(watermarked_image)
display_image(extracted_watermark)
```
